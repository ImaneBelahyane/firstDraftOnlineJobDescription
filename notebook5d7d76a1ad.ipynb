```python
import re
import fitz
import spacy
from spacy.lang.fr import French
from spacy.lang.en import English
import unicodedata
def normalize(text):
    return unicodedata.normalize('NFKD', text.casefold()).encode('ASCII', 'ignore').decode().strip()



# Define regular expressions for section headings
tools_regex = re.compile(r'Tools|Technologies|Skills')
education_regex = re.compile(r'Education|Qualifications')
work_experience_regex = re.compile(r'Work Experience|Experience')
language_regex = re.compile(r'Language|Languages')
activities_regex = re.compile(r'Activities|Interests|skills')

# Initialize empty dictionaries for each section
tools_dict = {}
education_dict = {}
work_experience_dict = {}
language_dict = {}
activities_dict = {}

# Open the PDF file in binary mode
with open('/kaggle/input/cv-samples/cv.pdf', 'rb') as pdf_file:
    # Initialize the PDF reader object
    pdf_document = fitz.open(stream=pdf_file.read(), filetype='pdf')

    # Iterate over each page of the PDF document
    for page in pdf_document:
        # Extract the text from the current page
        text = page.get_text()

        # Normalize the text
        text = normalize(text)

        # Search for sections in the text of the current page using regular expressions
        if tools_regex.search(text):
            # Extract the tools section
            tools_dict['tools'] = text[tools_regex.search(text).end():]
        if education_regex.search(text):
            # Extract the education section
            education_dict['education'] = text[education_regex.search(text).end():]
        if work_experience_regex.search(text):
            # Extract the work experience section
            work_experience_dict['work_experience'] = text[work_experience_regex.search(text).end():]
        if language_regex.search(text):
            # Extract the language section
            language_dict['language'] = text[language_regex.search(text).end():]
        if activities_regex.search(text):
            # Extract the activities section
            activities_dict['activities'] = text[activities_regex.search(text).end():]

# Convert the dictionaries to dataframes
tools_df = pd.DataFrame.from_dict(tools_dict, orient='index')
education_df = pd.DataFrame.from_dict(education_dict, orient='index')
work_experience_df = pd.DataFrame.from_dict(work_experience_dict, orient='index')
language_df = pd.DataFrame.from_dict(language_dict, orient='index')
activities_df = pd.DataFrame.from_dict(activities_dict, orient='index')

# Concatenate the dataframes along the columns axis
result = pd.concat([tools_df, education_df, work_experience_df, language_df, activities_df], axis=1)

# Display the result
print(result)
```
