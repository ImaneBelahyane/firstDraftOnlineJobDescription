```python
import pandas as pd
import unicodedata
import fitz
import re

def normalize(text):
    return unicodedata.normalize('NFKD', text.casefold()).encode('ASCII', 'ignore').decode().strip()

# Initialize empty dictionaries for each section

from langdetect import detect

# Define translations of keywords in different languages
translations = {
    'fr': {
        'informatiques': 'informatiques',
        'formations': 'formations',
        'experiences': 'expériences',
        'langue': 'langue',
        'membre': 'membre'
    },
    'en': {
        'informatiques': 'computing',
        'formations': 'education',
        'experiences': 'work experience',
        'langue': 'language',
        'membre': 'activities'
    }
    # Add more translations for other languages as needed
}

# Detect the language of the text
language = detect(text)

tools_dict = {}
education_dict = {}
work_experience_dict = {}
langue_dict = {}
activities_dict = {}

# Open the PDF file in binary mode
with open('/kaggle/input/cv-samples/cv.pdf', 'rb') as pdf_file:
    # Initialize the PDF reader object
    pdf_document = fitz.open(stream=pdf_file.read(), filetype='pdf')

    # Iterate over each page of the PDF document
    for page in pdf_document:
        # Extract the text from the current page
        text = page.get_text()

        # Normalize the text
        text = normalize(text)

        # Search for sections in the text of the current page
        if 'informatiques' in text:
            # Extract the tools section
            tools_dict['tools'] = text[text.find('informatiques'):]
        if 'formations' in text:
            # Extract the education section
            education_dict['education'] = text[text.find('formations'):]
        if 'experiences' in text:
            # Extract the work experience section
            work_experience_dict['work_experience'] = text[text.find('experiences'):]
        if 'langue' in text:
            # Extract the langue section
            langue_dict['langue'] = text[text.find('langue'):]
        if 'membre' in text:
            # Extract the activities section
            activities_dict['activities'] = text[text.find('membre'):]

# Convert the dictionaries to dataframes
tools_df = pd.DataFrame.from_dict(tools_dict, orient='index')
education_df = pd.DataFrame.from_dict(education_dict, orient='index')
work_experience_df = pd.DataFrame.from_dict(work_experience_dict, orient='index')
langue_df = pd.DataFrame.from_dict(langue_dict, orient='index')
activities_df = pd.DataFrame.from_dict(activities_dict, orient='index')

# Concatenate the dataframes along the columns axis
result = pd.concat([tools_df, education_df, work_experience_df, langue_df, activities_df], axis=1)

# Display the result
print(result)
```

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import SnowballStemmer
```

```python
def process_text(text):
    # Tokenize the text
    tokens = word_tokenize(text)
    # Remove stopwords
    tokens = [token for token in tokens if token not in stopwords.words('english')]
    # Stem the words using SnowballStemmer
    stemmer = SnowballStemmer('english')
    tokens = [stemmer.stem(token) for token in tokens]
    # Remove duplicates
    tokens = list(set(tokens))
    # Join the tokens back into a string
    processed_text = ' '.join(tokens)
    return processed_text
```
